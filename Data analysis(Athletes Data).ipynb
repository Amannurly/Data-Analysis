```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
data = pd.read_csv('file:///C:/Users/Admin/Downloads/archive/athlete_events.csv')
data.head()
```

```python
reg_data = pd.read_csv('file:///C:/Users/Admin/Downloads/archive/noc_regions.csv')
reg_data.head()
```

```python
#Join the dataframes
athletes_data = data.merge(reg_data, how='left', on ='NOC')
athletes_data.head()
```

```python
# Age Distributions of the participants
plt.figure(figsize=(12,6))
plt.title('Age Distribution of the athletes')
plt.xlabel('Age')
plt.ylabel('Number of participartions')
plt.hist(athletes_data['Age'],np.arange(10,80,2), color='orange', edgecolor='white');
```

```python
#Male and Female participants
gender_counts = athletes_data['Sex'].value_counts()
gender_counts
```

```python
plt.figure(figsize=(12,6))
plt.title('Gender Distribuution')
plt.pie(gender_counts, labels = gender_counts.index, autopct='%.2f', startangle=150,shadow=True);
```

```python
# Total medals
athletes_data['Medal'].value_counts()
```

```python
#Total number of female athletes in each summer olympics
female_participants = athletes_data[(athletes_data['Sex']=='F') & (athletes_data['Season']=='Summer')][['Sex','Year']]
female_participants = female_participants.groupby('Year').count().reset_index()
female_participants.head()
```

```python
womenolympics = athletes_data[(athletes_data['Sex']=='F') & (athletes_data['Season']=='Summer')]
womenolympics.head()
```

```python
sns.set(style='darkgrid')
plt.figure(figsize=(12,6))
sns.countplot(x='Year',data=womenolympics,palette='Spectral')
plt.title('Woman participation');
```

```python
part = womenolympics.groupby('Year')['Sex'].value_counts()
plt.figure(figsize=(12,6))
part.loc[:,'F'].plot()
plt.title('Plot of Female Atheletes over time');
```

```python
# Gold medal athletes
gold_medals = athletes_data[athletes_data['Medal']=='Gold']
gold_medals.head()
```

```python
# Only the values that are different from NaN
gold_medals = gold_medals[np.isfinite(gold_medals['Age'])] 
gold_medals.head()
```

```python
#Gold beyond 60
gold_medals['ID'][gold_medals['Age']>60].count()
```

```python
sporting_event = gold_medals['Sport'][gold_medals['Age']>60]
sporting_event
```

```python
# Plot for sporting event
plt.figure(figsize=(10,5))
plt.tight_layout()
sns.countplot(sporting_event)
plt.title('Gold medals for athletes for over 60 years');
```

```python
# Gold medals from each country
totalgm = gold_medals['Region'].value_counts().reset_index(name='Medal').head()
totalgm
```

```python
g = sns.catplot(x='index', y ='Medal', data = totalgm, height=5, kind='bar', palette='rocket')
g.despine(left=True)
g.set_xlabels('Top 5 countries')
g.set_ylabels('Number of medals')
plt.title('Gold medals per Country');
```
