```python
# Create 2 new columns
epl_df['Mins/Match'] = (epl_df['Mins']/epl_df['Matches']).astype(int)
epl_df['Goals/Match'] = (epl_df['Goals']/epl_df['Matches']).astype(float)
epl_df
```

```python
#Total Goals
total_goals = epl_df['Goals'].sum()
total_goals
```

```python
# Penalty Goals 
pen_goals = epl_df['Penalty_Goals'].sum()
pen_goals
```

```python
# Penalty Attempts
total_penattem = epl_df['Penalty_Attempted'].sum()
total_penattem
```

```python
# Pie chart for penalties missed vs scored
plt.figure(figsize=(12,6))
pl_not_scored = total_penattem - pen_goals
data = [pl_not_scored, pen_goals]
labels = ['Penalty Missed', 'Penalty Scored']
color = sns.color_palette('Set2')
plt.pie(data, labels=labels, colors=color, autopct='%.0f%%')
plt.show()
```

```python
# Unique positions
epl_df['Position'].unique()
```

```python
# Total Fw players
epl_df[epl_df['Position']=='FW']
```

```python
# Players from different nations
np.size(epl_df['Nationality'].unique())
```

```python
# Most players from which countries
nationality = epl_df.groupby('Nationality').size().sort_values(ascending=False)
nationality.head(10).plot(kind='bar', figsize=(12,6), color=sns.color_palette('magma'));
```

```python
# Clubs with maximum players in their squad
epl_df['Club'].value_counts().nlargest(5).plot(kind='bar', figsize=(12,6), color=sns.color_palette('viridis'));
```

```python
#Clubs with least players in their club
epl_df['Club'].value_counts().nsmallest(5).plot(kind='bar', figsize=(12,6),color=sns.color_palette('viridis'));
```

```python
# Players based on age group
under20 = epl_df[epl_df['Age']<=20]
age20_25 = epl_df[(epl_df['Age']>20) & (epl_df['Age']<=25)]
age25_30=epl_df[(epl_df['Age']>25) & (epl_df['Age']<=30)]
above30=epl_df[epl_df['Age']>30]
```

```python
x = np.array([under20['Name'].count(),age20_25['Name'].count(),age25_30['Name'].count(),above30['Name'].count()])
labels=['<=20','21<=x<=25','26<=x<=30','x>=31']
plt.title('Total players with Age groups',fontsize=20)
plt.pie(x,labels=labels, autopct='%.1f%%')
plt.show()
```

```python
# Total under 20 palyers in each club
players_under_20 = epl_df[epl_df['Age']<20]
players_under_20['Club'].value_counts().plot(kind='bar', figsize=(12,6), color=sns.color_palette('magma'));
```

```python
# Under 20 players in Arsenal 
players_under_20[players_under_20['Club']=='Arsenal']
```

```python
# Under 20 players in Man city
players_under_20[players_under_20['Club']=='Manchester City']
```

```python
# Average age of players in each club
plt.figure(figsize=(12,6))
sns.boxplot(x='Club', y='Age', data=epl_df)
plt.xticks(rotation=90);
```

```python
num_player=epl_df.groupby('Club').size()
data = (epl_df.groupby('Club')['Age'].sum())/num_player
data.sort_values(ascending=False)
```

```python
# Total assists from each club
assists = pd.DataFrame(epl_df.groupby('Club', as_index=False)['Assists'].sum())
sns.set_theme(style="whitegrid")
ax = sns.barplot(x='Club', y='Assists', data=assists.sort_values(by='Assists'),palette='Set2')
ax.set_xlabel('Club')
ax.set_ylabel('Assists')
plt.xticks(rotation=90)
plt.rcParams['figure.figsize']=(20,8)
plt.title('Plot of Club and Total Assists')
```

```python
# Top 10 assists
top_10_assistans = epl_df[['Name','Club','Assists','Matches']].nlargest(n=10,columns='Assists')
top_10_assistans
```

```python
# Most goals by players
top_10_goaleadors = epl_df[['Name','Club','Goals','Matches']].nlargest(n=10,columns='Goals')
top_10_goaleadors
```

```python
# Goals per match
top_player_goalspermatch=epl_df[['Name','Goals/Match','Matches','Goals']].nlargest(n=10,columns='Goals/Match')
top_player_goalspermatch
```

```python
# Pie Chart - Goals with assists and without assists
plt.figure(figsize=(12,6))
assists = epl_df['Assists'].sum()
data=[total_goals-assists,assists]
labels=['Goals without assists','Goals with assists']
color = sns.color_palette('Set1')
plt.pie(data,labels=labels,colors=color,autopct='%.0f%%')
plt.show()
```

```python
# Top 10 players with most yellow cards
epl_yellow = epl_df.sort_values(by='Yellow_Cards',ascending=False)[:10]
plt.figure(figsize=(12,6))
plt.title('Top 10 players with most yellow cards')
c= sns.barplot(x=epl_yellow['Name'],y=epl_yellow['Yellow_Cards'],label='Players',color='yellow')
plt.ylabel('Number of Yellow Cards')
c.set_xticklabels(c.get_xticklabels(),rotation=45)
c;
```
