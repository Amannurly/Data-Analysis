```python
genre=[]
for x in df['Genre']:
    for y in x.split(','):
        genre.append(y.strip().lower())
genre
```

```python
from collections import Counter
```

```python
count=Counter(genre)
count
```

```python
count=count.most_common()
count
```

```python
top_genres = pd.DataFrame(count,columns=['Genre','Count'])
top_genres
```

```python
fig,ax=plt.subplots(figsize=(20,7))
ax.set(facecolor='Black')
sns.barplot(x=top_genres['Count'],y=top_genres['Genre'],palette='hls')
plt.title('Top Genres',fontweight='bold',fontsize=15)
plt.xlabel('Movies')
plt.ylabel('Genre')
plt.show()
plt.tight_layout()
```
